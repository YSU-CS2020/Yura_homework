## Coding Practice

```python
num = -3
num_sign = ("Positive" if num > 0 else ("Negative" if num < 0 else "Zero"))
assert num_sign == "Negative"
print("Question 4 passed!")
```

```python
flag = "hello"
flag_type = (
    "Integer" if isinstance(flag, int)
    else "String" if isinstance(flag, str)
    else "Boolean" if isinstance(flag, bool)
    else "Other"
)
assert flag_type == "String"
print("Question 5 passed!")
```

```python
value = 100
match type(value):
    case int:
        value_type = "Integer"
    case str:
        value_type = "String"
    case bool:
        value_type = "Boolean"
    case _:
        value_type = "Other"
assert value_type == "Integer"
print("Question 6 passed!")
```

```python
score = 95
if score >= 90:
    grade = "A"
elif 80 <= score < 90:
    grade = "B"
elif 70 <= score < 80:
    grade = "C"
else:
    grade = "F"
assert grade == "A"
print("Question 7 passed!")
```

```python
score = 82
match score:
    case s if s >= 90:
        grade = "A"
    case s if s in range(80, 90):
        grade = "B"
    case s if s in range(70, 80):
        grade = "C"
    case _:
        grade = "F"
assert grade == "B"
print("Question 8 passed!")
```

```python
day = "Friday"
match day.lower():
    case "monday" | "tuesday" | "wednesday" | "thursday" | "friday":
        day_type = "Weekday"
    case "saturday" | "sunday":
        day_type = "Weekend"
    case _:
        day_type = "Unknown"
assert day_type == "Weekday"
print("Question 9 passed!")
```

```python
a, b, op = 9, 3, "/"
match op:
    case "+":
        answer = a + b
    case "-":
        answer = a - b
    case "*":
        answer = a * b
    case "/":
        answer = a // b if b else "Error: Division by zero"
assert answer == 3
print("Question 21 passed!")
```
